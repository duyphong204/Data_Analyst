## Import thư viện cần thiết

```python
import pandas as pd
from ydata_profiling import ProfileReport
```

## Load Dataset

```python
marketing_data = pd.read_csv("marketing_campaign.csv")
```

## Báo cáo phân tích dữ liệu Marketing 

```python
# Tạo báo cáo phân tích tự động (profiling report) cho DataFrame marketing_data
# Báo cáo này sẽ tổng hợp thông tin như: số lượng giá trị thiếu, phân bố dữ liệu,
# mối tương quan giữa các cột, điểm ngoại lai, loại dữ liệu,...
profile = ProfileReport(marketing_data)

# Xuất báo cáo ra file HTML có tên "profile_output.html"
# Sau khi chạy xong, mở file này trong trình duyệt để xem báo cáo chi tiết
profile.to_file("profile_output.html")
```
