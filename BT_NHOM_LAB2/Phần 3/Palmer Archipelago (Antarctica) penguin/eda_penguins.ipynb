# Phân tích dữ liệu hai biến trên dữ liệu về chim cánh cụt.

## Import một số thư viện cần thiết

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Load Dataset và giữ lại những cột quan trọng


```python
# Đọc dữ liệu từ file CSV "penguins_size.csv" vào DataFrame penguins_data
penguins_data = pd.read_csv("penguins_size.csv")

# Giữ lại 4 cột cần thiết: species (loài), culmen_length_mm (chiều dài mỏ),
# body_mass_g (khối lượng cơ thể, gram), và sex (giới tính)
penguins_data = penguins_data[['species', 'culmen_length_mm', 'body_mass_g', 'sex']]
```

## Biểu đồ phân tán (Scatter Plot) giữa Chiều dài mỏ và Khối lượng cơ thể

```python
# case 1: basic
# Vẽ biểu đồ scatter plot cơ bản giữa chiều dài mỏ và khối lượng cơ thể
sns.scatterplot(
    data = penguins_data,                     # Dữ liệu sử dụng
    x = penguins_data["culmen_length_mm"],    # Trục hoành (chiều dài mỏ)
    y = penguins_data["body_mass_g"]          # Trục tung (khối lượng cơ thể)
)

# case 2: advanced
plt.figure(figsize=(12, 6))  # Tạo khung hình (figure) với kích thước 12x6 inch

# Vẽ scatter plot nâng cao, cho phép tùy chỉnh thêm bằng đối tượng trục (ax)
ax = sns.scatterplot(
    data = penguins_data,                     # Dữ liệu
    x = penguins_data["culmen_length_mm"],    # Trục X
    y = penguins_data["body_mass_g"]          # Trục Y
)

# Đặt tiêu đề cho biểu đồ, với kích thước chữ là 20
ax.set_title(
    'Bivariate analysis of Culmen Length and Body Mass',  # Tiêu đề biểu đồ
    fontsize = 20                                         # Cỡ chữ
)
```

```python
# Tạo bảng chéo (cross-tabulation) thể hiện số lượng chim cánh cụt theo loài và giới tính
pd.crosstab(
    index = penguins_data['species'],   # Biến hàng: loài chim (species)
    columns = penguins_data['sex']      # Biến cột: giới tính (sex)
)
```

```python
# Tạo bảng tổng hợp (pivot table) tính chiều dài mỏ trung bình của từng loài chim cánh cụt
pd.pivot_table(
    penguins_data,               # Dữ liệu nguồn
    values='culmen_length_mm',   # Cột cần tính toán (chiều dài mỏ)
    index='species',             # Nhóm theo loài chim
    aggfunc=np.mean              # Hàm thống kê: trung bình
)
```

```python
# Vẽ ma trận biểu đồ phân tán (pair plot) giữa các biến trong bộ dữ liệu penguins
sns.pairplot(
    data = penguins_data    # Bộ dữ liệu đầu vào
)
```
